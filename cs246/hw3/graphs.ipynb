```python
import csv
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
data = []

for i in [1,2,4,8]:
    filename = 'cacti-sweep.out'
    with open(filename) as csvfile:
        readCSV = csv.reader(csvfile, delimiter=',')
        for row in readCSV:
            data.append(row)

xrow = [int(x) for x in np.transpose(data)[0]]
yrow = [y for y in np.transpose(data)[1]]
xrow=xrow[:3]
yrow=yrow[:3]

fig, ax = plt.subplots(figsize=(18,10))
ax.plot(xrow, yrow, linestyle='--', marker='o', color='b')
ax.set_ylabel("Access Time (ns)",fontsize=20)
ax.set_xlabel("Associativity (1-4)",fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
ax.grid()
ax.set_title('CACTI: Cache Access Time (ns) vs. Associativity (1-4)', fontsize=18)
plt.savefig('cacti_assoc_sweep.png')
```

```python
data = []

#for j in range(1,7):
for i in [1,2,4,8]:
    filename = 'quantum/out/cache_assoc_{}.out'.format(str(i))
    with open(filename) as csvfile:
        readCSV = csv.reader(csvfile, delimiter=',')
        for row in readCSV:
            row = list(map(float, row))
            data.append(row)

xrow = [int(x) for x in np.transpose(data)[0]]
yrow = [100 * y for y in np.transpose(data)[3]]

fig, ax = plt.subplots(figsize=(18,10))
ax.plot(xrow, yrow, linestyle='--', marker='o')
ax.set_ylabel("L1DCache Misses (%)",fontsize=20)
ax.set_xlabel("Associativity",fontsize=20)
ax.set_ylim([10,25])
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
ax.grid()
plt.axhline(y=23.75, linestyle='--')
ax.legend(['pin simulation', 'perf'])
ax.set_ylim([0,25])
ax.set_title('libquantum: Cache Misses (%) vs. Associativity (Direct-Mapped L1DCache)', fontsize=18)
plt.savefig('quantum_assoc_1_novic.png')
```
